```python
import pandas as pd
import seaborn as sb
import numpy as np

from matplotlib import pyplot as plt
```

```python
sb.set(font_scale=2) 
sb.set_style("whitegrid")
fig, ax = plt.subplots(
    nrows=1,
    ncols=3,
    sharey='row',
    sharex='col',
    figsize=(20,8)
)

fig.subplots_adjust(wspace=0.08) # hspace=0

color = ['darkred', 'y', 'seagreen']
use_case = ("A", "B", "C")
y_pos = np.arange(len(use_case))
ax[0].set_ylabel('Fehler-Codes in %')

# 1500
error_percent_1500 = [0, 0.0435, 0]

ax[0].set_xlabel("Anwendungsfall")
ax[0].set_title('1500 VU')
ax[0].bar(y_pos, error_percent_1500, align='center', alpha=1, color=color)
ax[0].set_xticks(range(len(use_case)))
ax[0].set_xticklabels(use_case)

# 2000
error_percent_2000 = [0.0071, 0.1362, 0.0237]

ax[1].set_xlabel("Anwendungsfall")
ax[1].set_title('2000 VU')
ax[1].bar(y_pos, error_percent_2000, align='center', alpha=1, color=color)
ax[1].set_xticks(range(len(use_case)))
ax[1].set_xticklabels(use_case)

# 4000
#error_percent_4000 = [1.3140, 5.6801, 1.7515]

#ax[2].set_xlabel("Anwendungsfall")
#ax[2].set_title('4000 VU')
#ax[2].bar(y_pos, error_percent_4000, align='center', alpha=1, color=color)
#ax[2].set_xticks(range(len(use_case)))
#ax[2].set_xticklabels(use_case)

plt.show()
```
