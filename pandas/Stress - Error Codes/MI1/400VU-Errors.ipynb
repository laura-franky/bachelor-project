```python
 def get_time_series(dataframe: pd.DataFrame) -> pd.Series:
    number_of_rows = len(dataframe.index)

    return pd.Series(
        name='time',
        data=[element * 10 for element in range(number_of_rows)]
    )
```

```python
def get_error_codes(time: pd.Series, dataframe: pd.DataFrame) -> pd.DataFrame:
    codes = pd.concat([time, dataframe["codes"]], axis=1)
    codes = codes.set_index("time")
    
    for key, val in codes["codes"].items():
        print(val)
    
        

        

    for key, value in codes['codes'][0].items():
        codes[key] = codes['codes'].apply(lambda object: object[key])

    codes = codes.drop(columns=["codes"])
    return codes
```

```python
def get_mean_rps(time: pd.Series, dataframe: pd.DataFrame) -> pd.DataFrame:
    rps = pd.concat([time, dataframe["rps"]], axis=1)
    rps = rps.set_index("time")

    for key, value in rps['rps'][0].items():
        rps[key] = rps['rps'].apply(lambda object: object[key])
    
    rps = rps.drop(columns=["count", "rps"])
    return rps
```

```python
time = get_time_series(A)
error_codes_a = get_error_codes(time, A)
mean_rps_a = get_mean_rps(time, A)
```
